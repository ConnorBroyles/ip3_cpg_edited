# Question 5-7

```python
car_sales.info()
```

```python
car_sales['Price']=car_sales['Price'].str.replace('$', '',regex=True)
car_sales['Price']=car_sales['Price'].str.replace(',', '',regex=True)
```

```python
car_sales.head(15)
```

```python
car_sales.info()
```

```python
car_sales['Price']=car_sales['Price'].astype(str).astype('int64')
```

```python
car_sales.info()
```

```python
car_sales.head()
```

```python
car_sales['Total Sales']=car_sales['Price'].cumsum(axis=0)
```

```python
car_sales.head()
```

# Question 8-15

```python
plt.plot(car_sales["Price"],car_sales["Total Sales"] )
plt.title("Price vs total sales line plot")
plt.xlabel('Price')
plt.ylabel('Total Sales')


plt.show()
```

```python
plt.scatter(car_sales["Odometer (KM)"],car_sales["Price"] )
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')


plt.show()
```

```python
plt.bar(car_sales["Make"], car_sales["Odometer (KM)"])
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')

plt.show()
```

```python
plt.hist(car_sales["Make"])
plt.xlabel("Make")

plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,ncols=2,figsize= (11,11))

ax1.plot(car_sales["Price"],car_sales["Total Sales"] )
ax1.set_title("Price vs total sales line plot")
ax1.set_xlabel('Price')
ax1.set_ylabel('Total Sales')


ax2.scatter(car_sales["Odometer (KM)"],car_sales["Price"] )
ax2.set_title("Odometer vs Price scatter plot plot")
ax2.set_xlabel('Odometer (KM)')
ax2.set_ylabel('Price')


ax3.bar(car_sales["Make"], car_sales["Odometer (KM)"])
ax3.set_title("Make vs Odometer bar plot")
ax3.set_xlabel('Make')
ax3.set_ylabel('Odometer (KM)')


ax4.hist(car_sales["Make"])
ax4.set_title('Histogram of car make')
ax4.set_xlabel("Make")


plt.show()
```

```python
box_plot=sns.boxplot(x='Colour',y='Total Sales',data=car_sales)
```

```python
lm_plot = sns.lmplot(x="Total Sales", 
                     y="Odometer (KM)", 
                     data=car_sales, 
                     height=5, 
                     hue='Make', 
                     ci=False )
```

```python
sns.pairplot(car_sales, plot_kws={'color':'green'})
plt.show()
```

```python
pc = car_sales[['Total Sales','Price','Odometer (KM)']].corr(method ='pearson')

cols = ['Total Sales','Price','Odometer (KM)']

ax = sns.heatmap(pc, annot=True,
                 yticklabels=cols,
                 xticklabels=cols,
                 annot_kws={'size': 60})
# Same as Seaborn heatmap on moodle
```
